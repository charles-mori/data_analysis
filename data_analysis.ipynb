```python

# 1.1 Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Use seaborn style
sns.set_style("whitegrid")

# 1.2 Load dataset
try:
    iris = load_iris()
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
    print("Dataset loaded successfully!")
except Exception as e:
    print("Error loading dataset:", e)

# 1.3 Display first few rows
df.head()
```

```python

# 1.4 Explore structure
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# 1.5 Clean dataset (if needed)
df = df.dropna()
print("\nDataset after cleaning:")
print(df.info())
```

```python

# 2.1 Basic statistics
df.describe()

# 2.2 Group by species and calculate mean for each group
grouped = df.groupby('species', observed=True).mean()
print("\nMean values grouped by species:\n", grouped)
```


**Observations:**

- Setosa generally has smaller sepal and petal sizes compared to Versicolor and Virginica.

- Virginica tends to have the largest petal length and petal width.

- Clear differences exist between species based on measurements.


```python

# 3.1 Line Chart - Trend over samples for petal length
plt.figure(figsize=(10,6))
plt.plot(df.index, df['petal length (cm)'], label='Petal Length', color='green')
plt.title('Petal Length Trend across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python

# 3.2 Bar Chart - Average Petal Length per Species
plt.figure(figsize=(8,6))
grouped['petal length (cm)'].plot(kind='bar', color=['purple', 'blue', 'orange'])
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python

# 3.3 Histogram - Distribution of Sepal Width
plt.figure(figsize=(8,6))
plt.hist(df['sepal width (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python

# 3.4 Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species', palette='Set2')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```


**Final Observations:**

- No missing values were found in the dataset.

- Virginica flowers have larger petals compared to Setosa and Versicolor.

- The sepal width is most frequently around 3 cm.

- The scatter plot clearly shows that species can be separated visually based on sepal and petal measurements.
